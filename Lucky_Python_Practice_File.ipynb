```python
# palindrome :- number equal to its reverse 
num = int(input("enter any number"))
temp = num 
reverse= 0 

while num>0:
    last = num%10
    reverse = reverse * 10 + last 
    num = num//10
    

print(reverse)
if reverse == temp:
    print("palindrome")
else:
    print("not palindrome")
```

```python
# amrstrong number :- 3 digit 



num = int(input("enter any number"))
temp = num 
sum= 0 

while num>0:
    last = num%10
    sum = sum + (last **3)
    num = num//10
    

print(sum)
if sum == temp:
    print("armstrong")
else:
    print("not armstrong ")
```

```python
# strong number 

# factorial 

# n= 5
# fact = 1
# while n>=1:
#     fact = fact * n
#     n-=1

print(fact)

num = int(input("enter any number"))
temp = num 
sum= 0 

while num>0:
    last = num%10
    
    n= last
    fact = 1
    while n>=1:
       fact = fact * n
       n-=1
    
    sum = sum + fact
    num = num//10
    

print(sum)
if sum == temp:
    print("strong")
else:
    print("not strong ")
```

```python
# perfect number
num = int(input("enter any number"))

i = 1 
sum = 0

while i<num:
    if num%i==0:
        print(i)
        sum+=i
    i+=1


print("sum of factors = ",sum)
if sum == num:
    print("perfect")
else:
    print("not perfect")
```

```python
# Prime Number 
num = int(input("enter any number"))

i = 1
sum = 0

while i<num:
    if num%i==0:
        print(i)
        sum+=i
    i+=1


print("sum of factors = ",sum)
if sum == 1:
    print("Prime")
else:
    print("Not Prime")
```
