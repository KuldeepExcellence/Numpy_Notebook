```python
# What is Linear Regression?
# Linear Regression is the simplest form of Machine Learning algorithm used for predicting a 
# continuous value (like salary, house price, temperature, etc.) based on one or more input variables.

# It assumes there’s a linear relationship between input (X) and output (Y).

```

```python
# Formula
# For Simple Linear Regression (one independent variable):
# Y=mX+c
#Y → Predicted value (dependent variable)
#X → Input value (independent variable)
#m → Slope of the line (how much Y changes when X increases by 1)
#c → Intercept (value of Y when X = 0)
```

```python
import pandas as pd
df = pd.read_csv("employee_data_with_salary.csv")
df.head()
```

```python
x= df[["Years_of_Experience"]]
y= df["Salary_k"]
```

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=99)
print(x_train,x_test,y_train,y_test)

```

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(x_train,y_train)
```

```python
y_pred = model.predict(x_test)
y_pred
```

```python
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error

r2 = r2_score(y_test,y_pred)
mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
print("Model Performance:")
print("R² Score (Accuracy): ",r2)
print("Mean Absolute Error: ",mae)
print("Mean Squared Error:", mse)

print("Slope (m):",model.coef_[0])
print("Intercept (c):", model.intercept_)


```

```python
import matplotlib.pyplot as plt

plt.scatter(x_test,y_test,color="b",label="Actual Salary")
plt.xlabel("Years of Experience")
plt.ylabel("Salary in K")
plt.legend()
```
