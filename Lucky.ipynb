```python
# find the factorial of a number using recursion .

# 6! = 6*5*4*3*2*1


# Recusrion. 
def fact(num):
    
    if num == 1 :
        return 1
    else:
        return num * fact(num-1)
        #   3 * fact(2)
              #   2* fact(1)
                        # 1 
    


num = int(input("enter any number")) 
ans = fact(num)
print("factorial = ",ans)




```

```python

# Name Mangling.

class Employee:
    def __init__(self):
        self.name = "aman"
        self.department = "tech"
        self.__salary=45000


    def getter(self,password):
        if password=="root@123":
            return self.__salary
        else:
            print("Incorrect password ")
    def setter(self,a,password):
        if password == "root@123":
            self.__salary=a
            return "Value Updated"
        else:
            print("Incorrect password ")


    def show_details(self):
        print(self.name)
        print(self.department)


obj = Employee()
print(obj.getter("root@12"))
print(obj.setter(65000,"root@12"))






#
# class A:
#     __a="hello"
#
#
# obj = A()
#
# print(obj.a)

```

```python
# Abstraction
# abstract class 
# abstract method
#       ABC = Abstract Base Class 
from abc import ABC , abstractmethod


class Car(ABC):
    @abstractmethod
    def chassynumber(self):
        pass
    def engine(self):
        print("800 CC engine")
    def gearbox(self):
        print("5 speed manual gear box ")

class Maruti(Car):
    def chassynumber(self):
        print("this is a a1 car")

class Tata(Car):
    def chassynumber(self):
        print("this is a a1 car")
    


car = Maruti()
car.engine()


cart= Tata()

    
        




```
