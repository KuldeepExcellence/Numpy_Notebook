```python
# Keyword Arguments 

def power(base , pow):

    c = base ** pow
    print(c)


power(pow = 10,base = 3)

# pow = keyword 
# 10 :- value 


# base = keyword 
# 3 = value 
```

```python
# Keyword Arbitrary Arguments .

def data(**a):
    print(a)
    for i in a.values():
        print(i)
    for j in a.keys():
        print(j)
    for k in a.items():
        print(k)

data(name="kuldeep",city="delhi",mobile="778877887788",course="python")

```

```python
# find the factorial of a number using recursion .

# 6! = 6*5*4*3*2*1


# Recusrion. 
def fact(num):
    
    if num == 1 :
        return 1
    else:
        return num * fact(num-1)
        #   3 * fact(2)
              #   2* fact(1)
                        # 1 
    


num = int(input("enter any number")) 
ans = fact(num)
print("factorial = ",ans)




```
